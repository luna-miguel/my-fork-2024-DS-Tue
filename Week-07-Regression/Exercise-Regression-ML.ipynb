````python
# Playoffs vs. Wins

NBA["PTSdiff"] = NBA["PTS"] - NBA["oppPTS"]
sns.regplot(data=NBA, x="PTSdiff", y="W", scatter_kws={'alpha': 0.25})
````

````python
independent_variables = "PTSdiff"
dependent_variable = 'W'

X = NBA[independent_variables]
y = NBA[dependent_variable]

X = sm.add_constant(X)

model_wins = sm.OLS(y, X).fit()
y_pred = model_wins.predict(X) 

results = model_wins.summary()
print(results)

# W = 41.0012 + (0.0326 * PTSdiff), which means 
# PTSdiff must be greater or equal to 30.67 in order to win 42 games
````

````python
# Points scored

independent_variables = ["2PA", "3PA", "FTA", "AST", "ORB", "DRB", "TOV", "STL", "BLK"]
dependent_variable = 'PTS'

X = NBA[independent_variables]
y = NBA[dependent_variable]

X = sm.add_constant(X)

model_wins = sm.OLS(y, X).fit()
y_pred = model_wins.predict(X) 

results = model_wins.summary()
print(results)

model_wins_root_mean_squared_error = rmse(y, y_pred)
print('\n', '='*79, '\n')
print( 'Dependent Variable %s average: ' % dependent_variable, y.mean())
print( 'RMSE:', model_wins_root_mean_squared_error)

# RSME: average difference between values predicted by a model and the actual values
````

````python
# Defensive rebounds, turnovers, and blocks are insignificant

independent_variables = ["2PA", "3PA", "FTA", "AST", "ORB"]
dependent_variable = 'PTS'

X = NBA[independent_variables]
y = NBA[dependent_variable]

X = sm.add_constant(X)

model_wins = sm.OLS(y, X).fit()
y_pred = model_wins.predict(X) 

results = model_wins.summary()
print(results)

model_wins_root_mean_squared_error = rmse(y, y_pred)
print('\n', '='*79, '\n')
print( 'Dependent Variable %s average: ' % dependent_variable, y.mean())
print( 'RMSE:', model_wins_root_mean_squared_error)

# Adj. R-squared remains relatively the same, so we are justified in removing these variables.
# RMSE slightly increases, but is negligible
````
